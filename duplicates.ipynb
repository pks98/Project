```python
import pandas as pd
```

```python
df=pd.read_csv('co2_emissions (1).csv',sep=';')
df.head()
```

```python
df.shape
```

```python
df['make']=df['make'].apply(lambda x:x.lower())
```

```python
cat_col=[]
for i in df.columns:
  if df[i].dtype == 'object':
    cat_col.append(i)
```

```python
cat_col
```

```python
for i in cat_col:
  df[i]=df[i].apply(lambda x:x.lower())
```

```python
df.head()
```

```python
df[df.duplicated()] 
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True) 
```
